````python
"""
Q5. Given: `sales = [12000, 15000, 11000, 18000, 17000]`

**Using a loop:**

1. Calculate total revenue.
2. Count how many months had sales above 15000.
3. Find maximum sales without using built-in max().
"""

sales = [12000, 15000, 11000, 18000, 17000]
total_revenue = 0
count_high_sales = 0
max_sale = sales[0]

for s in sales:
    total_revenue += s
    if s > 15000:
        count_high_sales += 1
    if s > max_sale:
        max_sale = s
print(f"Total Revenue: {total_revenue}")

print(f"Months with sales > 15000: {count_high_sales}")

print(f"Maximum Sales: {max_sale}")
````

2. Explain difference between List comprehension and Generator expression.

   -> A **List comprehension** creates the entire list in memory at once (e.g., `[x*x for x in range(10)]`), whereas a **Generator expression** (e.g., `(x*x for x in range(10))`) yields items one by one on demand.

3. Why are generators preferred in big data scenarios?

   -> Generators are preferred in big data because they are "memory efficient." They do not store the entire sequence in RAM, which is critical when processing millions or billions of records that would otherwise cause an "Out of Memory" error.

````python
"""
Q10. Design a class:
class Product:
With:
● Attributes: name, price, quantity
● Method: calculate_revenue()
Create 3 objects and print their revenue.
"""
# Designing the Product class
class Product:
    def __init__(self, name, price, quantity):
        # Initializing attributes
        self.name = name
        self.price = price
        self.quantity = quantity

    # Method to calculate revenue
    def calculate_revenue(self):
        return self.price * self.quantity

# Create 3 objects (instances) of the Product class
p1 = Product("Laptop", 50000, 2)
p2 = Product("Smartphone", 25000, 4)
p3 = Product("Headphones", 1500, 10)

# Print their revenue
print(f"Revenue for {p1.name}: {p1.calculate_revenue()}")
print(f"Revenue for {p2.name}: {p2.calculate_revenue()}")
print(f"Revenue for {p3.name}: {p3.calculate_revenue()}")
````
