Q1. Write a SQL query to display:
- customer name
- city
- signup date

Sort results by signup_date (latest first).

 ->

    SELECT name, city, signup_date
    FROM Customers
    ORDER BY signup_date DESC;


Q2. Find all orders placed in the last 30 days where:
- total_amount > 5000

  Sort by highest order value first.

->

    SELECT *
    FROM Orders
    WHERE order_date >= CURRENT_DATE - INTERVAL '30 days'
     AND total_amount > 5000
    ORDER BY total_amount DESC;


Q3. Explain:

a) What is the relational model?

-> A method of structuring data using tables (relations) with rows and columns. It relies on the logic that data is related through common attributes.

b) Why are primary keys and foreign keys essential?

->
- Primary Keys uniquely identify a record in a table, ensuring no duplicates.
- Foreign Keys link tables together, maintaining referential integrity between related data.

c) What problems does normalization solve?

-> Solves data redundancy (repeating info) and update anomalies (where changing data in one place leaves it incorrect in another).

d) When might denormalization be preferred in analytics systems?

-> Preferred in analytics to improve read performance. By combining tables ahead of time, you reduce the number of complex joins needed for heavy reporting.

Q4. Compare and contrast:

- Subqueries
1. When each should be used - Filtering results based on another query.
2. Performance - Can be slow if "correlated" (runs per row).
3. Readability - Can become messy if nested deeply.

- Joins
1. When each should be used - Combining rows from 2+ tables.
2. Performance - Generally fastest; optimized by engines.
3. Readability - Good for simple relations.

- CTEs
1. When each should be used - Breaking down complex logic into steps.
2. Performance - Similar to subqueries (depends on engine).
3. Readability - Excellent; creates a logical flow.

Q6. Use the following simplified schema:

    Customers(customer_id, name, city, signup_date)
    Orders(order_id, customer_id, order_date, total_amount)
    Order_Items(order_id, product_id, quantity, price)
    Products(product_id, product_name, category, cost_price, selling_price)

Write a query to calculate:

- Total revenue per product category

- Total quantity sold

  Sort by highest revenue first.

->

    SELECT
      p.category,
      SUM(oi.quantity * oi.price) AS total_revenue,
      SUM(oi.quantity) AS total_quantity_sold
    FROM Products p
    JOIN Order_Items oi ON p.product_id = oi.product_id
    GROUP BY p.category
    ORDER BY total_revenue DESC;

Q8. Write a query to:
- Identify duplicate customers based on same name and city
- Show number of duplicate occurrences

Explain how you would clean such data.
->
    SELECT name, city, COUNT(*) as occurrence_count
    FROM Customers
    GROUP BY name, city
    HAVING COUNT(*) > 1;


Q9.

 a) What is a View?

   -> A virtual table that runs its query every time you call it. It uses no disk space but can be slow for complex logic.


 b) How is it different from a Materialized View?

  ->  Actually stores the result on disk.

 c) When should analytics teams prefer materialized views?

  -> Teams prefer Materialized Views for massive datasets where the underlying data doesn't change every second, as it provides near-instant results.

d) What are the risks of overusing views?

  -> Overusing views can lead to "View on View" nesting, which becomes a nightmare to debug and causes massive performance hits.
