Q 3. A hospital wants to:
● Predict patient readmission risk
● Automate report generation
● Deploy a chatbot for patient FAQs

Map each problem to:
● Traditional Data Science
● Machine Learning
● GenAI
● Agentic AI
Explain your reasoning.

->
- Predicting readmission risk = Machine Learning because you have labeled historical data (patient was/was not readmitted) and you train a model to find the patterns that predict the label. This is classic supervised learning.

- Automating report generation = Traditional Data Science because it mainly involves querying, aggregating, and visualizing existing structured data. No "learning" is strictly needed - just well-defined analytics, statistics, and templating.

- Patient FAQ chatbot = GenAI because it must generate coherent, context-appropriate natural-language responses. Large Language Models are designed exactly for this generative text capability.

- Agentic AI would come into play if you wanted an autonomous system that plans and executes multiple steps on its own - for example, an AI that monitors incoming patients, automatically flags high-risk ones, generates the report, and notifies a doctor, all without step-by-step human instructions.

```python
"""
Q 7.You are given a customer review:
review = "The product quality is Excellent but delivery was Late"
Write Python code to:
1. Convert to lowercase
2. Count number of words
3. Check if the word "excellent" exists
4. Replace "Late" with "delayed"
"""
# Convert to lowercase
review_lower = review.lower()
print(review_lower)

# Count number of words
word_count = len(review.split())
print("Number of words:", word_count)

# Check if the word "excellent" exists (case-insensitive)
exists = "excellent" in review.lower()
print('Is "excellent" present?', exists)

# Replace "Late" with "delayed"
updated_review = review.replace("Late", "delayed")
print(updated_review)

```

```python
"""
Q8.Given sales data:
sales = [12000, 15000, 11000, 18000, 17000]
1. Calculate total revenue.
2. Find average revenue.
3. Identify highest sales month.
4. Append new month sale = 20000.
5. Sort the list in descending order.

"""
sales = [12000, 15000, 11000, 18000, 17000]
months = ["Month1", "Month2", "Month3", "Month4", "Month5"]

# 1. Total revenue
total_revenue = sum(sales)
print("Total Revenue:", total_revenue)

# 2. Average revenue
average_revenue = sum(sales) / len(sales)
print("Average Revenue:", average_revenue)

# 3. Highest sales month
highest = max(sales)
highest_month = months[sales.index(highest)]
print("Highest Sales:", highest, "in", highest_month)

# 4. Append new month sale = 20000
sales.append(20000)
print("After append:", sales)

# 5. Sort in descending order
sales.sort(reverse=True)
print("Sorted (descending):", sales)
```

```python
"""
Q9.You are building a simple sales analysis system.
Given:
products = ["Laptop", "Phone", "Tablet"]
prices = [50000, 20000, 15000]
quantities = [5, 10, 7]
Tasks:
1. Compute total revenue for each product.
2. Identify which product generated maximum revenue.
3. Store revenues in a new list.
4. Print a formatted summary like:
"""
products = ["Laptop", "Phone", "Tablet"]
prices = [50000, 20000, 15000]
quantities = [5, 10, 7]

revenues = []
for i in range(len(products)):
    revenues.append(prices[i] * quantities[i])

print(revenues)   # [250000, 200000, 105000]

max_product = products[revenues.index(max(revenues))]

for i in range(len(products)):
    print(f"{products[i]}: Revenue = ₹{revenues[i]}")

print(f"Highest Revenue: {max_product} (₹{max(revenues)})")
```

Q 10.A startup collects customer feedback as text and numerical ratings.
Explain step-by-step:

1. How raw text becomes structured data.
   - Raw text is unstructured and messy. It becomes structured through preprocessing, where we clean the text (removing punctuation), tokenize it (splitting sentences into words), and then convert those words into numbers (vectors). Finally, these numbers are stored in a table with columns like "Customer ID" and "Sentiment Score."

2. Where Python (strings & lists) is used.
   - Python Strings are used for text cleaning, such as using .lower() to normalize text or .replace() to fix typos. Lists are used to store collections of data, such as a list of all customer reviews or a list of individual words extracted from a sentence for analysis.

3. Where Data Science models are used.
   - Models are used for Sentiment Analysis to automatically label feedback as positive, negative, or neutral. They are also used for Topic Modeling, which automatically groups thousands of reviews into categories like "Pricing," "Shipping," or "Product Quality."

4. Where GenAI can enhance the system.
   - GenAI can summarize thousands of long reviews into a short, readable paragraph for managers. It can also auto-generate personalized, polite responses to customer complaints, saving time for the support team.

5. What ethical concerns must be addressed.
   - The biggest concerns are Data Privacy (removing customer names before analysis) and Algorithmic Bias (ensuring the AI doesn't misinterpret certain languages or slang). There must also be Transparency, so customers know their feedback is being processed by an AI system.
